#### Тест-кейсы

WSSG = Web Site Section Goods

##### Тест-кейс №1 (Позитивный)

> ID: WSSG-1

> Заголовок: Проверить корректность добавления товаров в корзину

> Предусловия: Необходимо выбрать определённые товары и добавить их в корзину

> Шаги:

* Перейти в раздел "Магазин"
* Выбрать товар
* Добавить товар в корзину
* Проверить наличие товара после перезагрузки страницы

> Постусловия:

> Ожидаемый результат: Товары сохранятся в корзине после перезагрузки

> Фактический результат: Товары сохранились в корзине после перезагрузки

> Статус: Success

> Комментарий: Так как и в других маркетплейсах есть возможность добавить отсутствующий товар в корзину, не посчитали это недочётом 

##### Тест-кейс №2 (Негативный)

> ID: WSSG-2

> Заголовок: Проверка валидности полей оформления заказа

> Предусловия: Добавить товары в корзину и провести оформление заказа; Вводимые данные будут на скриншоте ниже (рис 1.1.)

> Шаги:

* Добавить товары в корзину
* Перейти в окно "Оформить заказ"
* Ввести данные
* Проверить, получим ли ошибку на некорректность введённых данных

> Постусловия:

> Ожидаемый результат: Получение ошибки

> Фактический результат: Отсутствие ошибки, продолжение оформления заказа
